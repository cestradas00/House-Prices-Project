```python
dff = df [['Id', 'OverallCond', 'LotShape', 'MSZoning', 'YrSold', 'LotArea', 'SalePrice']]
dff
```

```python
dff.isna().sum()
```

#### 1. Relationship between Lot  Area Vs Sale Price

```python
dff1 = dff.loc[:, ['LotArea','SalePrice', 'YrSold']]
```

```python
sns.scatterplot(data=dff1, x='LotArea', y='SalePrice', hue='YrSold')
plt.title('Scatter Plot (LotArea vs SalePrice)')
plt.show()
```

For this case we can notice that there are some outliers in the plot, so proceed to extract them.

```python
z = np.abs(stats.zscore(dff1['LotArea']))
dff1 = dff1[z < 3]
```

```python
sns.scatterplot(data=dff1, x='LotArea', y='SalePrice', hue='YrSold')
plt.title('Scatter Plot (LotArea vs SalePrice)')
plt.show()
```

```python
dff1[['LotArea', 'SalePrice']].corr()
```

The correlation shows us that there is a positive correlation between Sale Price and Lot Area but it is not strong.

#### 2. Relationship between Overall Cond vs Sale Price

```python
dff['OverallCond'].value_counts()
```

```python
sns.boxplot(data=dff.loc[:, ['OverallCond','SalePrice']], x='OverallCond', y='SalePrice')
plt.title('Box Plot of SalePrice by OverallCond')
plt.show()
```

```python
dff.groupby('OverallCond')['SalePrice'].agg(['mean', 'median']).transpose()
```

In this plot we can observe that the most expensive houses do not seem to be directly related to the overall conditions, the two most expensive groups are the 5 and the 9. It's important to mention that group 5 has the most sales by far.

#### 3. Relationship between Sale Prices and MSZoning through YearsSold

```python
sns.barplot(data = dff, x='YrSold', y= 'SalePrice', hue='MSZoning', errorbar=None)
plt.show()
```

In this plot, we can notice that the sale prices were maintained similar through the years of selling, also, we can observe that the MSZones are constant with the prizes, with the exception of C(all) in 2007.

#### 4. Relationship between Lot Shape vs Lot Area

```python
dff['LotShape'].value_counts()
```

```python
sns.boxplot(data=dff.loc[:, ['LotShape','LotArea']], x='LotShape', y='LotArea')
plt.title('Box Plot of LotShape by LotArea')
plt.show()
```

```python
dff.groupby('LotShape')['LotArea'].agg(['mean', 'median']).transpose()
```

In the graph, we can see that the regular lots are the smallest. Also, we can notice that the outliers are affecting the mean in the IR2 and IR3 types of lots.
